Contains a spacetime diagram to answer a question about relativity from a curious correspondent.

```python
import pandas as pd 
import plotnine as p9

%matplotlib inline
```

```python
(
    p9.ggplot(pd.DataFrame([], columns=['X', 'T']), p9.aes(x='X', t='T'))
    
    + p9.xlim(-3, 3)
    + p9.ylim(-1, 4)
    + p9.theme_void()
    + p9.theme(plot_background=p9.element_rect(fill='white'), figure_size=(10, 10))
    + p9.coord_fixed( ratio=1)
    + p9.geom_abline(slope=0, intercept=0) + p9.annotate("text", x=3, y=0.1, label="$T_0=0$")
    + p9.geom_abline(slope=0, intercept=1) + p9.annotate("text", x=3, y=1.1, label="$T_0=0.1$")
    + p9.annotate("point", x=0, y=0, size=5)
    + p9.annotate("point", x=0, y=1, size=5)
    + p9.annotate("point", x=1.73, y=2, size=8, fill='purple',color='purple')
    + p9.geom_abline(slope=-0.866, intercept=0, color='red') + p9.annotate("label", x=1, y=-0.9, label="$T_1=0$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=0.5, color='red') + p9.annotate("label", x=1, y=-0.4, label="$T_1=0.1$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=1.0, color='red') + p9.annotate("label", x=1, y=0.1, label="$T_1=0.2$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=1.5, color='red') + p9.annotate("label", x=2, y=-0.26, label="$T_1=0.3$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=2.0, color='red') + p9.annotate("label", x=2, y=0.24, label="$T_1=0.4$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=2.5, color='red') + p9.annotate("label", x=2, y=0.75, label="$T_1=0.5$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=3.0, color='red') + p9.annotate("label", x=2, y=1.25, label="$T_1=0.6$", color='red', angle=-40)
    + p9.geom_abline(slope=-0.866, intercept=3.5, color='red') + p9.annotate("label", x=2, y=1.75, label="$T_1=0.7$", color='red', angle=-40)
    
    + p9.geom_abline(slope=0.866, intercept=0, color='blue') + p9.annotate("label", x=-1, y=-0.9, label="$T_2=0$", color='blue', angle=40)
    + p9.geom_abline(slope=0.866, intercept=0.5, color='blue') + p9.annotate("label", x=-1, y=-0.4, label="$T_2=0.1$", color='blue', angle=40)
    + p9.geom_abline(slope=0.866, intercept=1.0, color='blue') + p9.annotate("label", x=-1, y=0.1, label="$T_2=0.2$", color='blue', angle=40)
    + p9.geom_abline(slope=0.866, intercept=1.5, color='blue') + p9.annotate("label", x=-1, y=0.6, label="$T_2=0.3$", color='blue', angle=40)
    + p9.geom_abline(slope=0.866, intercept=2.0, color='blue') + p9.annotate("label", x=-1, y=1.1, label="$T_2=0.4$", color='blue', angle=40)
    + p9.geom_abline(slope=0.866, intercept=2.5, color='blue') + p9.annotate("label", x=-1, y=1.6, label="$T_2=0.5$", color='blue', angle=40)
    + p9.geom_abline(slope=1.15,intercept=0, color="purple", size=2)
)
```
